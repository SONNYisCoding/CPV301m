```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([(1,2),(2,3),(3,4),(4,1)])
```

```python
plt.figure(figsize=(6,4))
nx.draw(G, with_labels=True, node_size = 1000)
plt.title('Ban đầu')
plt.show()
```

```python
G.add_node(5)
G.add_edge(4,5)
```

```python
plt.figure(figsize=(6,4))
nx.draw(G, with_labels=True, node_size = 1000)
plt.title('Sau khi thêm')
plt.show()
```

```python
G.remove_edge(4,3)
G.remove_node(2)
```

```python
plt.figure(figsize=(6,4))
nx.draw(G, with_labels=True, node_size = 1000)
plt.title('Sau khi xóa')
plt.show()
```
